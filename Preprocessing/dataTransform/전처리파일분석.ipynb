### 면적 데이터를 사용하기 위해 university_locations.csv 전처리

```python
import pandas as pd
df = pd.read_csv('university_locations.csv')
```

```python
df
```

```python
import pandas as pd

new_university = {
    '대학명': '경희대학교',
    '면적': '407,376㎡',
    '위도': 37.59685,
    '경도': 127.0518
}

new_university_df = pd.DataFrame([new_university])

df = pd.concat([df, new_university_df], ignore_index=True)

df['대학명'] = df['대학명'].str.replace(r'\[.*?\]', '', regex=True).str.strip()

# 전처리 작업
# 1. 숫자와 특수문자 제거
df['면적'] = df['면적'].str.replace('[^0-9]', '', regex=True)  # 면적에서 숫자만 남기기
df['면적'] = pd.to_numeric(df['면적'], errors='coerce')  # 숫자로 변환

# 2. 위도와 경도의 null 값 처리
# (null 값은 빈 문자열이나 0으로 변환하거나 제거 가능)
df['위도'] = pd.to_numeric(df['위도'], errors='coerce')  # 숫자로 변환, 잘못된 값은 NaN 처리
df['경도'] = pd.to_numeric(df['경도'], errors='coerce')

# NaN 값 확인 (필요에 따라 채우거나 삭제)
df['위도'] = df['위도'].fillna(0)  # 위도 NaN 값을 0으로 채우기
df['경도'] = df['경도'].fillna(0)  # 경도 NaN 값을 0으로 채우기


# 3. 컬럼 이름 Hive에 맞게 변경
df.rename(columns={
    '대학교': 'univ_name',
    '면적': 'square',
    '위도': 'latitude',
    '경도': 'longitude'
}, inplace=True)

```

### numData에는 전문대학교가 없는 걸 확인 - 수동으로 추가해줌

```python
# 학교 이름 정규화: 공백 제거 및 특수문자 제거
df['대학명'] = df['대학명'].str.strip().str.replace(r'[^\w가-힣]', '', regex=True)
numData['학교'] = numData['학교'].str.strip().str.replace(r'[^\w가-힣]', '', regex=True)

# df에는 있지만 numData에 없는 학교 찾기
missing_in_numData = df[~df['대학명'].isin(numData['학교'])]

# numData에는 있지만 df에 없는 학교 제거
numData = numData[numData['학교'].isin(df['대학명'])]

numData = numData.reset_index(drop=True)

# 결과 확인
print("numData에 없는 df의 학교:")
print(missing_in_numData)

print("\n정리된 numData:")
print(numData)
```

```python
new_data = {
    '학교': [
        '인덕대학교', '한양여자대학교', '한국폴리텍대학', '삼육보건대학교',
        '서일대학교', '동양미래대학교', '명지전문대학', '숭의여자대학교',
        '배화여자대학교', '서울여자간호대학교'
    ],
    '총재학생수': [8163, 7538, 2345, 1598, 8562, 9568, 9350, 3312, 3086, 960]
}

new_data_df = pd.DataFrame(new_data)

numData = pd.concat([numData, new_data_df], ignore_index=True)

# 결과 확인
print(numData)
```

```python
# numData와 df를 병합하여 '총재학생수' 컬럼 추가
df = df.merge(numData, left_on='대학명', right_on='학교', how='left')

# '학교' 컬럼은 중복되었으므로 삭제
df = df.drop(columns=['학교'])

df['총재학생수'] = pd.to_numeric(df['총재학생수'], errors='coerce').fillna(0).astype(int)

# 결과 확인
print(df)
print("\ndf.info():")
print(df.info())
```
